```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터 불러오기 (경로 수정 필요)
df = pd.read_csv("data/preprocessed.csv", header=None)

# label 칼럼 변환: retained -> 0, churned -> 1
df["label"] = df["label"].map({"retained": 0, "churned": 1})

# device 칼럼 변환: Android -> 1, iPhone -> 0
df["device"] = df["device"].map({"Android": 1, "iPhone": 0})

# device별 churn 비율 계산
device_churn = df.groupby("device")["label"].mean()

# device별 churn 비율 시각화
plt.figure(figsize=(8, 6))
sns.barplot(x=device_churn.index, y=device_churn.values, palette="coolwarm")
plt.xticks([0, 1], ["iPhone", "Android"])
plt.title("Churn Rate by Device Type")
plt.xlabel("Device Type")
plt.ylabel("Churn Rate")
plt.show()
```
